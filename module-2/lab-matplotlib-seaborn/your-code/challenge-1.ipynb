#### Import all the libraries that are necessary.

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.subplot(1,2,1)
plt.plot(y)
plt.subplot(1,2,2)
plt.plot(z)
plt.show()
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, ax = plt.subplots(1,2)
ax[0].plot(x,y)
ax[1].plot(x,z)
plt.show()
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.plot(y)
plt.subplot(1,2,2)
plt.plot(z)
plt.show()

fig, ax = plt.subplots(1,2,figsize=(15,4))
ax[0].plot(x,y)
ax[1].plot(x,z)
plt.show()
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
plt.figure(figsize=(15,4))
ax = plt.subplot(1,2,1)
plt.plot(y)
ax.set_xlabel("x")
ax.set_ylabel("y")
axis = plt.subplot(1,2,2)
plt.plot(z)
axis.set_xlabel("x")
axis.set_ylabel("z")
plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
plt.figure(figsize=(12,5))
ax = plt.subplot(1,2,1)
plt.plot(x**2)
plt.plot(e**x)
ax.set_xscale("linear")
ax.set_yscale("linear")


axis = plt.subplot(1,2,2)
plt.plot(x**2)
plt.plot(e**x)
axis.set_xscale("linear")
axis.set_yscale("log")
axis.set_ylim(ymin=1)
axis.set_xlim(xmin=0)
```

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
# your code here
df = pd.read_csv("vehicles.csv")
plt.scatter(df["Combined MPG"],df["Highway MPG"])
plt.xlabel("Combined MPG")
plt.ylabel("Highway MPG")
plt.show()
```

Box Whisker Plot

```python
# your code here
plt.boxplot(df["CO2 Emission Grams/Mile"])
plt.ylabel("CO2 Emission Grams/Mile")
plt.show()
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# your code here
plt.hist(df["Fuel Barrels/Year"])
plt.xlabel("Fuel Barrels/Year")
plt.ylabel("Number of cars")
plt.show()
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
df_group = (df[["Fuel Type","City MPG"]].groupby("Fuel Type").agg({"City MPG":"mean"})).reset_index()
```

```python
df_group
```

```python
plt.bar(df_group["Fuel Type"],df_group["City MPG"])
plt.xlabel("Fuel Type")
plt.ylabel("City MPG")
plt.xticks(range(len(df_group["Fuel Type"])),[e for e in df_group["Fuel Type"]], rotation=45, horizontalalignment="right")
plt.show()
```
