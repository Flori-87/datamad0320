# Challenge 1 - Analysis of Variance

In this part of the lesson, we will perform an analysis of variance to determine whether the factors in our model create a significant difference in the group means. We will be examining a dataset of FIFA players. We'll start by loading the data using the code in the cell below.

```python
# Run this code:

fifa = pd.read_csv('fifa.csv')
```

Let's examine the dataset by looking at the `head`.

```python
# Your code here:

fifa.head()
```

Player's values are expressed in millions of euros. We would like this column to be numeric. Therefore, let's create a numeric value column. Do this by stripping all non-numeric characters from each cell. Assign this new data to `ValueNumeric`. There is no need to multiply the value to be expressed in millions. 

```python
# Your code here:

def onlyNumber(value):
    pattern = r"\d+\.?\d*"
    return float((re.search(pattern,value)).group())

fifa["ValueNumeric"] = fifa["Value"].apply(onlyNumber)
fifa.head()
```

```python
#Renombro Preferred Foot como Foot para que no me de problemas en el modelo por tener dos palabras separadas
fifa = fifa.rename(columns={"Preferred Foot": "Foot"})
```

#### We'd like to determine whether a player's preffered foot and position have an impact on their value. 

Using the `statsmodels` library, we are able to produce an ANOVA table without munging our data. Create an ANOVA table with value as a function of position and preferred foot. Recall that pivoting is performed by the `C` function.

Hint: For columns that have a space in their name, it is best to refer to the column using the dataframe (For example: for column `A`, we will use `df['A']`).

```python
#para poder producir la tabla de ANOVA primero hay que generar el modelo de regresión lineal
# utilizo el “*” para además de incluir la relación de ValueNumeric con Foot y Position, que además incluya la relación 
# de ValueNumeric con respecto a Foot y Position juntos
mod = ols(formula='ValueNumeric ~ Foot * Position', data=fifa)
res = mod.fit()
print(res.summary())
sm.stats.anova_lm(res,typ=2)
```

What is your conclusion from this ANOVA?

### Your conclusions here:

Según el resultado del anova, el p-value es menor de 0.05 en el caso de la Posición, por lo que rechazamos la hipótesis nula en este caso, es decir, las medias de los valores de los futbolistas son diferentes entre sí si se tiene en cuenta la posición.
Sin embargo, si atendemos a la preferencia del pie o a la unión de la preferencia del pie y la posición, el p-value es mayor de 0.05, por lo que no podemos rechazar la hipotesis nula, es decir, para estos parametros la media de valores son iguales.
Esto cuadraría con el hecho de que F está cercano a 1 en estos dos ultimos casos, es decir, no hay mucha dispersión de los datos cuando atendemos a esos parámetros.
Por lo tanto, esto quiere decir que la posición sí afecta al valor de los futbolistas

After looking at a model of both preffered foot and position, we decide to create an ANOVA table for nationality. Create an ANOVA table for numeric value as a function of nationality.

```python
# Your code here:

mod = ols(formula='ValueNumeric ~ Nationality', data=fifa)
res = mod.fit()
print(res.summary())
sm.stats.anova_lm(res,typ=2)
```

What is your conclusion from this ANOVA?

como el p-value es menor de 0.05, rechazamos la hipotesis nula, es decir, las medias de los valores de los futbolistas no son iguales cuando atendemos a la nacionalidad. Es decir, la nacionalidad afecta al valor de los futbolistas, pero habría que mirar cada nacionalidad y ver el p-value de cada una

# Challenge 2 - Linear Regression

Our goal with using linear regression is to create a mathematical model that will enable us to predict the outcome of one variable using one or more additional independent variables.
